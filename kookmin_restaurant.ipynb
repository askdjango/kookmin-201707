```python
import json
import requests

params = {
    'callback': '_',
    'sdate': '2017-08-01',
    'edate': '2017-08-31',
}

menu_url = 'http://kmucoop.kookmin.ac.kr/menu/menujson.php'

jsonp_string = requests.get(menu_url, params=params).text
json_string = jsonp_string.replace('_(', '').replace(');', '')
menu_dict = json.loads(json_string)
menu_dict
```

```python
menu_dict.keys()
```

```python
for date, category_list in menu_dict['한울식당'].items():
    print(date)
    for name, menu in category_list.items():
        if menu['메뉴']:
            menu_names = ' '.join(menu['메뉴'].split())
            print(name, menu_names, menu['가격'])
    print()
```
